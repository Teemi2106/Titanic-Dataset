```python
import pandas as pd

# Load the data
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# Handling missing values
df.dropna(inplace=True)

```

```python
df.head()
```

```python
# Dropping irrelevant columns from the dataset
df = df.drop(columns=['Name', 'Ticket'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# Using Label Encoding to get unique values in the columns and fitting them
encoder = LabelEncoder()
df['Sex'] = encoder.fit_transform(df['Sex'])
df['Embarked'] = encoder.fit_transform(df['Embarked'])
df['Cabin'] = encoder.fit_transform(df['Cabin'])
```

```python
df.head()
```

```python
# Seperating the features and target variables
Y = df['Survived']
X = df.drop(columns='Survived')
```

```python
from sklearn.model_selection import train_test_split


# Splitting the data into training and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

# Initializing and training the LogisticRegression model
model = LogisticRegression(max_iter = 1000)
model.fit(X_train, Y_train)
```

```python
model
```

```python
# Making predictions with the fitted model
predictions = model.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Calculating accuracy of the model using different metrics 
accuracy = accuracy_score(Y_test, predictions)
precision = precision_score(Y_test, predictions)
recall = recall_score(Y_test, predictions)
f1 = f1_score(Y_test, predictions)
roc_auc = roc_auc_score(Y_test, model.predict_proba(X_test)[:, 1])

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("ROC AUC Score:", roc_auc)
```
